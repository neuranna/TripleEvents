```python
dir = '/Users/ruimingao/Documents/TripleEvents/TripleEvents_firstlevel_spm/040_EventsOrig'

import nilearn
import nilearn.image
```

```python
image_1_path = dir + '/con_0001.nii'
image_2_path = dir + '/con_0002.nii'
image_3_path = dir + '/con_0011.nii'
image_1 = nilearn.image.load_img(image_1_path)
image_2 = nilearn.image.load_img(image_2_path)
image_3 = nilearn.image.load_img(image_3_path)
image_1_data = image_1.get_fdata()
image_2_data = image_2.get_fdata()
image_3_data = image_3.get_fdata()
```

```python
# histogram 
img = image_1_data - image_2_data
import matplotlib.pyplot as plt
plt.hist(img.flatten(), bins=100)
plt.show()
```

```python
# histogram 
img2 = image_3_data
import matplotlib.pyplot as plt
plt.hist(img2.flatten(), bins=100)
plt.show()
```
